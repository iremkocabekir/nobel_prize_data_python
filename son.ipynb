```python
df12 = df11.groupby(["years", "sex"]).agg(count=("sex", "count")).reset_index()
df13 = df11.groupby(["years"]).agg(mean=("age", "mean")).reset_index()
plt.figure(figsize=(16, 6))
stacked_data_count = df12.pivot(index='years', columns='sex', values='count')
stacked_data_count.plot(kind='bar', stacked=True, color=['skyblue', 'orange'])
plt.twinx()
plt.plot(df13["years"], df13["mean"], color='green', marker='o')
plt.xlabel("years")
plt.ylabel("mean")
plt.xticks(rotation=45)
plt.legend(["mean", "Male", "Female"], loc='upper left')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
